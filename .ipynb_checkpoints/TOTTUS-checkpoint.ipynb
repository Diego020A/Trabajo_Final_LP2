```python
# FUNCIONES QUE UTILIZAREMOS MAS ADELANTE

from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import xml.etree.ElementTree as ET
import time
import pandas as pd


#------
def it():
    timeout=25
    element_present = EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/section/div[1]/section/div[2]/div[2]/ul[1]'))
    WebDriverWait(browser, timeout).until(element_present)
    elem_list = browser.find_element(
        By.XPATH, "/html/body/div[1]/section/div[1]/section/div[2]/div[2]/ul[1]")
    items = elem_list.find_elements(
        By.CLASS_NAME, 'product')
    return items


def alm():
    table = []
    for item in it():
        timeout=25
        element_present = EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/section/div[1]/section/div[2]/div[2]/ul[1]'))
        WebDriverWait(browser, timeout).until(element_present)
        prov = []
        
        pagina="TOTTUS"
        title = ""
        price = ""
        cantidad = ""
        
        prov.append(pagina)
        try:
            title = item.find_element(By.TAG_NAME, 'h2').text
            prov.append(title)
        except:
            pass
        
        try:
            cantidad = item.find_element(By.TAG_NAME, "h3").text
            prov.append(cantidad)
        except:
            pass
        
        try:
            precio = item.find_element(By.CLASS_NAME, 'list price medium cmrPrice'.replace(" ",".")).text
            prov.append(precio)
        except:
            pass
    
        table.append(prov)
    return table

print("hecho")
```
