```python
# PRUEBA FINAL (todos los productos de Metro)

# Se importan librerias
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import random
from time import sleep
from selenium.webdriver.chrome.service import Service

options = webdriver.ChromeOptions()
options.add_argument('--incognito')
s= Service('C:/Users/Diego/Desktop/Trabajo_Final_LP2/chromedriver.exe')   # Se utiliza comando Service() porque arrojaba error
                                                                          # deprecated .... iperkenel.
driver = webdriver.Chrome(service=s,options = options)

# Se asigna la pagina web a scrapear
driver.get("https://www.metro.pe/")  

# Se crean listas para la iteración de categorías
lista_menu=["frutas-y-verduras","carnes-aves-y-pescados","desayuno","lacteos","embutidos-y-fiambres",
      "abarrotes","aguas-y-bebidas"]
tipos=["frutas-y-verduras","carnes-aves-y-pescados","desayuno","lacteos","embutidos-y-fiambre",
       "comprar-abarrotes-online","aguas-y-bebidas--naturales-y-gaseosas"]

total_datos=[]

for i in lista_menu:
    boton1=driver.find_element("xpath",'//button[@class="btn red toggle-menu"]')
    boton1.click()  # Hace click para abrir la barra de categorías
    sleep(1)
    a='//div[@class="item frutas-y-verduras"]'
    a=a.replace("frutas-y-verduras",i)

    
    boton2= driver.find_element("xpath",a)  
    boton2.click()  # Hace click en la categoría a scrapear

    sleep(1)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")     # Hace scroll directo hacia el final de página
    sleep(5)
    WebDriverWait(driver, 120).until(
        EC.presence_of_all_elements_located((
            By.XPATH,'//div[@class="shelf-content"]/div[@style="display: none;"]')))  # Espera hasta que cargue la página

    # El total de productos se dividen en productos y últimos productos (last)
    
    # Iteración para la descarga de productos y últimos productos (last) de cada categoría
    for e in tipos:
        b='//li[@class="frutas-y-verduras"]'
        c='//li[@class="frutas-y-verduras last"]'
        b=b.replace("frutas-y-verduras",e)
        c=c.replace('frutas-y-verduras',e)
        elemento1= driver.find_elements("xpath",b)
        elemento2= driver.find_elements("xpath",c)
        tipos.pop(0)
        break

    sleep(1)
    # Analiza cada producto y extrae su nombre (con cantidad) y precio 
    
    # Extrae para productos
    for elem in elemento1:
        saco=[]
        try:
            precio= elem.find_element("xpath",'.//span[@class="product-prices__value product-prices__value--best-price"]').text
            nombre= elem.find_element("xpath",'.//a[@class="product-item__name"]').text
            print(nombre)
            print(precio)
            saco.append(nombre)
            saco.append(precio)
            total_datos.append(saco)
        except:
            precio = 'NO DISPONIBLE'
            producto = 'NO DISPONIBLE'
    
    # Extrae para últimos productos (last)
    for elem in elemento2:
        a=''
        b= ''
        saco=[]
        try:
            producto= elem.find_element("xpath",'.//a[@class="product-item__name"]').text
            precio= elem.find_element("xpath",'.//span[@class="product-prices__value product-prices__value--best-price"]').text
            if producto!=b:
                saco.append(producto)
                print(producto)
                if precio!=a:
                    saco.append(precio)
                    print(precio)    
               
            if saco[1]!=a or saco[0]!=b:
                F.append(saco)
        except:
            precio = 'NO DISPONIBLE'
            producto = 'NO DISPONIBLE'
#Cierre de página
driver.close()
```

```python
# LIMPIEZA Y ESTRUCTURACIÓN DE DATOS DESCARGADOS

# Se crea la columna Cantidad
# Se importan librerias
import re
import pandas as pd

total1=[]
saco1=[]      # Nombre y Precio
saco2=[]      # Cantidad

for i in total_datos:
    saco1=[]   # nombre-precio-cantidad
    saco2=[]   # cantidad
    saco3=[]   # Agrupa los datos de cada producto
    saco1.append(i[0])
    for e in i:
        if e==i[0]:
            x= re.findall('\sx\s',e)    # Posee letra _x_ ?
            y= re.findall('[Pp]ack',e) # Posee palabra P-pack en el texto ?
            z= re.findall('^Mix',e)     # Posee letra Mix al comienzo ?
            w=re.findall('\sx.+',e)
            o= re.findall('\s[/]\s',e)
            u= re.findall('\s[+]\s',e)
            q=re.findall('\d.+',e)
            oq=re.findall('\d+',e)
            if len(y)>0:
                saco2.append('Pack') #se almacena pack como cantidad
            elif len(z)>0:
                saco2.append('Mix') # se almacena Mix como cantidad

            elif len(x)>0:
                if len(w)>0:
                    saco2.append(w[0])

            elif len(x)==0:
                if len(oq)==0:
                    if len(u)>0:
                        saco2.append('Duo')
                    else:
                        saco2.append('1 unidad')
                elif len(o)>0:
                    saco2.append(q[0])
                elif len(u)==0:
                    v=re.findall('\d+\s+([Mm]l|[Kk]g|g|[Uu]nid|[Ll]t|[Ll]|[Ss]obres|gr|[Uu]n\.|[Uu]nidades|Filtrantes|[Ll]itro|[Ll]itros|[Uu]n|[Oo]nzas)',e)
                    v1=re.findall('\d+([Mm]l|[Kk]g|g|[Uu]nid|[Ll]t|[Ll]|[Ss]obres|gr|[Uu]n\.|[Uu]nidades|Filtrantes|[Ll]itro|[Ll]itros|[Uu]n|[Oo]nzas)',e)
                    # v y v1 sirven para encontrar condiciones en la oracion 'e'
                    if len(oq)>=2:
                        t2=re.findall('%',e)
                        lp=e.split()
                        a=0
                        for elem in lp:
                            t=re.findall('^\d.+([Mm]l|[Kk]g|g|[Uu]nid|[Ll]t|[Ll]|[Ss]obres|gr|[Uu]n\.|[Uu]nidades|Filtrantes|[Ll]itro|[Ll]itros|[Uu]n|[Oo]nzas)$',elem)
                            t1=re.findall('^([Mm]l|[Kk]g|g|[Uu]nid|[Ll]t|[Ll]|[Ss]obres|gr|[Uu]n\.|[Uu]nidades|Filtrantes|[Ll]itro|[Ll]itros|[Uu]n|[Oo]nzas)$',elem)

                            if len(t)==1:
                                saco2.append(lp[a])
                            elif len(t1)==1:
                                saco2.append(lp[a-1]+' '+lp[a])
                            a=a+1
                    else:
                        saco2.append(q[0])
                        
                    if len(v)==0 or len(v1)==0:
                        saco2.append(q[0])
                    
                elif len(u)>0:
                    saco2.append('Pack')
            else:
                saco2.append(q[0])
            
            if saco2[0]=='Pack' or saco2[0]=='Mix':
                saco3.append('Metro')
                saco3.append(saco1[0])
            else:
                saco1=saco1[0].replace(saco2[0],'')
                saco3.append('Metro')
                saco3.append(saco1)
                
        elif e==i[1]:
            e=e.split()
            for q in e:
                z=re.findall('\d.+',q)
                if len(z)>0:
                    saco3.append(z[0])
            
    saco3.append(saco2[0])
    total1.append(saco3)        

## Longitud de los datos
# len(total1

# Se crea dataframe con datos ya estructurados con las columnas Nombre-Precio(S/.)-Cantidad

columnas=['Local','Nombre','Precio','Cantidad']
datos1=pd.DataFrame(data=total1,columns=columnas)           
datos1
```

```python
# Envío de de los resultados al correo ingresado anteriormente (SIN COMPARAR)

from pretty_html_table import build_table
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
import getpass

# Se establece un usuario y contraseña por defecto quien va a enviar el correo
# usuario: jossemen01@gmail.com
# contraseña: mxszphkwuhexmfue       <- contraseña generado para aplicaciones autorizadas en la cuenta gmail quien envía.

sender_address = 'jossemen01@gmail.com'
sender_pass = 'mxszphkwuhexmfue'
receiver_address = correo

# El cuerpo del mensaje puede ser html, json, texto puro
body = """
<html>
<head>
</head>

<body>
        {0}
</body>

</html>
""".format(build_table(resultados, 'blue_light'))    # Se puede cambiar el estilo de la tabla 'blue_light'


message = MIMEMultipart("alternative")  # Tipos de mensaje multiparte como html
message['From'] = sender_address
message['To'] = receiver_address
message['Subject'] = 'Resultados de la comparación de producto en 3 localidades'   

message.attach(MIMEText(body, 'html'))   # plain -> texto puro(plano)   html-> formato html  json...
# Establece el tipo de correo y su puerto (587)
session = smtplib.SMTP('smtp.gmail.com', 587)
# Inicia sesión
session.starttls()
session.login(sender_address, sender_pass)
text = message.as_string()
# Envío de correo
session.sendmail(sender_address, receiver_address, text)
session.quit()

print('Los resultados de la comparación fueron enviados con éxito a su correo')
```
