```python
# No se puede sacar tercera columna de cantidades (2 nombres en un producto -> genera error)

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import random
from time import sleep
from selenium.webdriver.chrome.service import Service

options = webdriver.ChromeOptions()
options.add_argument('--incognito')
s= Service('C:/Users/Diego/Desktop/Trabajo_Final_LP2/chromedriver.exe')   # Se utiliza comando Service() porque arrojaba error
                                                                          # deprecated .... iperkenel.
driver = webdriver.Chrome(service=s,options = options)

driver.get("https://www.metro.pe/")  

#confirmacion= driver.find_element("xpath",'//button[@class="swa12-confirm swa12-styled"]')
#confirmacion.click()

lista_menu=["frutas-y-verduras","carnes-aves-y-pescados","desayuno","lacteos","embutidos-y-fiambres",
      "abarrotes","aguas-y-bebidas"]
tipos=["frutas-y-verduras","carnes-aves-y-pescados","desayuno","lacteos","embutidos-y-fiambre",
       "comprar-abarrotes-online","aguas-y-bebidas--naturales-y-gaseosas"]

total_datos=[]

for i in lista_menu:
    boton1=driver.find_element("xpath",'//button[@class="btn red toggle-menu"]')
    boton1.click()
    sleep(2)
    a='//div[@class="item frutas-y-verduras"]'
    a=a.replace("frutas-y-verduras",i)

    
    boton2= driver.find_element("xpath",a)
    boton2.click()

    sleep(2)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    sleep(5)
    WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((
            By.XPATH,'//span[@class="product-prices__value product-prices__value--best-price"]')))
    
    #70 -> desayuno 10-> frutas,carnes  40->lacteos   25->embutidos 55->abarrotes 20-> agua
    if i=="desayuno":
        sleep(85)
    elif i=="abarrotes":
        sleep(70)
    elif i=="lacteos":
        sleep(40)
    elif i=="embutidos-y-fiambres":
        sleep(25)
    elif i=="aguas-y-bebidas":
        sleep(20)
    elif i=="frutas-y-verduras" or i=="carnes-aves-y-pescados":
        sleep(12)
    
    for e in tipos:
        b='//li[@class="frutas-y-verduras"]'
        b=b.replace("frutas-y-verduras",e)
        elemento= driver.find_elements("xpath",b)
        tipos.pop(0)
        break

    sleep(2)

    for elem in elemento:
        saco=[]
        try:
            precio= elem.find_element("xpath",'.//span[@class="product-prices__value product-prices__value--best-price"]').text
            nombre= elem.find_element("xpath",'.//a[@class="product-item__name"]').text
            print(nombre)
            print(precio)
            saco.append(nombre)
            saco.append(precio)
            total_datos.append(saco)
            #dict_t={'Frutas':{'precio':precio, 'producto':producto}}
            #datos.append(dict_t)
        except:
            precio = 'NO DISPONIBLE'
            producto = 'NO DISPONIBLE'
            
    sleep(3) # duda en sacar este tiempo
```

```python
#limpieza de datos (Se agrega columna cantidad)

total=[]
for i in total_datos:
    saco1=[]
    saco2=[] #cantidad
    for e in i:
        x=re.findall('\d.+',e)  # cantidad en texto y numero en precio
        #print(x[0])
        if e==i[0]:
            y=re.findall('\D+',e)  # nombre en texo
            saco1.append(y[0])   # nombre
            z=re.findall('\sx\s',e)
            if len(z)>0:
                o=re.findall('\sx.+',e)
                saco2.append(o[0])
            elif len(z)==0:
                if len(x)==0:
                    saco2.append('1 unidad')
                else:
                    saco2.append(x[0])   # cantidad

        elif e==i[1]:
            z=re.findall('\sx\s',e)
            if len(z)>=0:
                o=re.findall('\d+[.]\S+',e)
                saco1.append(o[0])
            
    #print(saco2[0])
    saco1.append(saco2[0])
    total.append(saco1)
total
# Problemas con los packs y 2 cantidades en mismo nombre de producto
#for i in total:
#    for e in i:
#        x=re.findall('[P-p]ack',e)
#        if len(x)>0:
#            print(i)
```
